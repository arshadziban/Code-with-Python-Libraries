```python
import pandas as pd
data = {
    'A': [10, 20, 30, 40, 50],
    'B': [15, 25, 35, 45, 55],
    'C': ['foo', 'bar', 'baz', 'qux', 'quux']
}
df = pd.DataFrame(data)
df
```

```python
#Iterating Over Rows in a Pandas DataFrame
for row in df.itertuples(index=True, name='Pandas'):
  print(row)
```

```python
#Selecting Rows Based on Conditions
filtered_df = df[df['A'] > 20]
filtered_df
```

```python
#Select Row Using iloc[] and iat[]
first_row = df.iloc[0]
first_row
```

```python
specific_value = df.iat[2,1]
specific_value
```

```python
#Limited Rows Selection with Specific Columns
limited_df = df.loc[:2, ['A', 'B']]
limited_df
```

```python
#Drop Rows Based on a Condition
drop_row = df[df['A'] <= 30]
drop_row
```

```python
#Randomly Select Rows
random_rows = df.sample(n=2)
random_rows
```

```python
# Create a List from Rows
row_list = df.values.tolist()
row_list
```
